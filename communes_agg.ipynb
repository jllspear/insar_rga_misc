```python
import geopandas as gpd
import pandas as pd
from shapely import wkt
import pyogrio
from tqdm import tqdm
import numpy as np

communes_path = "C:/Users/eleonore.kong/Documents/InSAR/communes.geojson"
data_communes_path = "C:/Users/eleonore.kong/Documents/InSAR/Liste des communes_20241129.csv"
```

```python
df_com = pyogrio.read_dataframe(communes_path)
data_com = pd.read_csv(
    data_communes_path,
    sep=';',
    decimal=',',
    dtype={
        0: str,  # Set column 0 to string
        3: str   # Set column 3 to string
    }
)
```

```python
data_com
```

```python
data = data_com.merge(df_com[['insee_com','geometry']], left_on='Code INSEE de la commune', right_on='insee_com')
data = gpd.GeoDataFrame(data, geometry='geometry',crs=4326)
```

```python
data
```

```python
data_f = data[(data['Groupements auxquels adhère la commune'].str.contains('CC')) | (data['Groupements auxquels adhère la commune'].str.contains('CA')) | data['Groupements auxquels adhère la commune'].str.contains('Métropole') | data['Groupements auxquels adhère la commune'].str.contains('Communauté') | data['Groupements auxquels adhère la commune'].str.contains('CU') | data['Groupements auxquels adhère la commune'].str.contains('Agglomération') | data['Groupements auxquels adhère la commune'].str.contains('Interco')]
data_f = data_f[~data_f['Groupements auxquels adhère la commune'].str.contains('SI')]
data_f = data_f[~data_f['Groupements auxquels adhère la commune'].str.contains('Syndicat')]
data_f = data_f[~data_f['Groupements auxquels adhère la commune'].str.contains('SIAEP')]
data_f = data_f[~data_f['Groupements auxquels adhère la commune'].str.contains('SIVOM')]
data_f = data_f[~data_f['Groupements auxquels adhère la commune'].str.contains('SIVU')]
data_f = data_f[~data_f['Groupements auxquels adhère la commune'].str.contains('SM')]

data_f
```

```python
pyogrio.write_dataframe(data_f, "C:/Users/eleonore.kong/Documents/InSAR/Communes_groupements.gpkg", driver='GPKG',layer='communes')
```
